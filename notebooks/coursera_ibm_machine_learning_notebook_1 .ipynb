```python

import pandas as pd

import numpy as np

import seaborn as sns

import matplotlib.pyplot as plt

sns.set_theme()
```

```python

df = sns.load_dataset("iris")

df.head()
```

```python

plt.plot(
    
    df.sepal_length,
    
    df.sepal_width,
    
    ls = "", 
    
    marker = "o",
    
    color = "darkblue"

)

plt.show()
```

```python

# -- plot 1

plt.plot(
    
    df.sepal_length,
    
    df.sepal_width,
    
    ls = "", 
    
    marker = "o",
    
    color = "crimson",
    
)

# -- plot 2

plt.plot(
    
    df.petal_length,
    
    df.petal_width,
    
    ls = "", 
    
    marker = "o",
    
    color = "purple",
    
)

plt.show()
```

```python

plt.hist(
    
    df.sepal_length, 
    
    bins = 25,
    
    color = "darkblue"

)

plt.show()
```

```python
fig, ax = plt.subplots()
```

```python

sns.jointplot(
    
    x = df["sepal_length"],
    
    y = df["sepal_width"],
    
    kind = "hex",
    
    color = "darkblue"
)

plt.show()
```

```python

sns.pairplot(
    
    df,
    
    hue = "species",
    
    palette = "ocean_d"
)

plt.show()
```

```python

df.groupby("species").mean().plot(color = ["red","blue","black","green"], fontsize = 10.0, figsize = (10,4))

plt.show()
```

```python

plot = sns.FacetGrid(df, col = "species", margin_titles = True)

plot.map(plt.hist, "sepal_width", color = "green")

plot = sns.FacetGrid(df, col = "species", margin_titles = True)

plot.map(plt.hist, "sepal_length", color = "darkblue")

plt.show()
```
